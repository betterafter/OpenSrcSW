```python
# 실습 12 - 1 : 서울의 일교차
import csv
 
f = open('seoul.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

date = []

for row in data:
    date.append( [  row[0], float(row[3]), float(row[4]), float(row[4]) - float(row[3]) ]  )
        


date.sort(key=lambda x:x[3])
b = len(date) - 1

print("일교차가 가장 큰 날")
print(date[b][0])
print('\n')

date.sort(key=lambda x:x[1])
b = 0

print("최저기온이 가장 낮은 날")
print(date[b][0])
print('\n')


date.sort(key=lambda x:x[2])
b = len(date) - 1

print("최고기온이 가장 높은 날")
print(date[b][0])
print('\n')

f.close() 
```

```python
# 실습 12 - 2 : 지하철 이용객
import csv

f = open('tmoney2.csv')

data = csv.reader(f)
header = next(data)

def string_to_int(str):
    res = ''
    for i in str:
        if (i >= '0' and i <= '9'):
            res += i
    return int(res)
    



d = dict()

for row in data:
    if row[1] in d:
        d[row[1]].append(  [row[3], int(string_to_int(row[4]) + string_to_int(row[5]))]  )
    else:
        d[row[1]] = []


for dd in d:
    print('\n\n\n')
    print("----------",dd,"----------")
    #print(d[dd])
    d[dd].sort(key=lambda x:x[1])
    b1 = int(len(d[dd])) - 1
    b2 = int(len(d[dd])) - 2
    b3 = int(len(d[dd])) - 3

    print("---이용객이 가장 많은 역 순위---")
    print(d[dd][b1][0],"(",d[dd][b1][1], ")")
    print(d[dd][b2][0],"(",d[dd][b2][1], ")")
    print(d[dd][b3][0],"(",d[dd][b3][1], ")")
    print('\n')
    print("---이용객이 가장 적은 역 순위---")
    print(d[dd][0][0],"(",d[dd][0][1], ")")
    print(d[dd][1][0],"(",d[dd][1][1], ")")
    print(d[dd][2][0],"(",d[dd][2][1], ")")

#print(max3)




f.close()
```
