```python
#week13 - 실습 1 : 생일 기온 시각화 
import matplotlib.pyplot as plt
import csv

f = open('week13_seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

high = []
low = []
for row in data:
    if(row[-1] != '' and row[-2] != ''):
        if(row[0].split('.')[1] == '4' and row[0].split('.')[2] == '15'):
            high.append(float(row[-1]))
            low.append(float(row[-2]))

plt.title('seoul')
plt.plot(low, color = 'skyblue', label = 'skyblue')
plt.plot(high, color = 'pink', label = 'pink')
plt.legend()
plt.show()

```

```python
#week13 - 실습 2 : 최고 기온 시각화 
import matplotlib.pyplot as plt
import csv

f = open('week13_seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

month_1 = []
month_8 = []
for row in data:
    if(row[-1] != ''):
        month = row[0].split('.')[1]
        if(month == '1'):
            month_1.append(float(row[-1]))
        elif(month == '8'):
            month_8.append(float(row[-1]))

plt.title('seoul')
plt.hist(month_8, bins=100, color='r')
plt.hist(month_1, bins=100, color='b')

plt.show()

```
